```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
df = pd.read_csv('pivoted_tensorboard_data.csv')

# Specify exactly which metrics you want to plot
metrics_to_plot = ['train_loss/implicit_loss']

plt.figure(figsize=(12, 8))

for metric in metrics_to_plot:
    if metric in df.columns:
        plt.plot(df['step'][0:560], df[metric][0:560], label=metric)
        plt.plot(df['step'][0:560], df['step'][0:560], label=metric)
plt.xlabel('Step')
plt.ylabel('Value')
plt.title('Training loss for equilibrium propagation approach on one molecule')
plt.legend()
plt.grid(True)
# plt.savefig('eq_prop.png', dpi=300, bbox_inches='tight')
plt.show()
```
