```python
from tbparse import SummaryReader

# Read from two different directories
reader1 = SummaryReader('./jacobian_1132', pivot=True)
reader2 = SummaryReader('./eq_prop', pivot=True)

df1 = reader1.scalars
df2 = reader2.scalars

# Save both datasets
df1.to_csv('experiment1_data.csv', index=False)
df2.to_csv('experiment2_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load both CSV files
df1 = pd.read_csv('experiment1_data.csv')
df2 = pd.read_csv('experiment2_data.csv')

plt.figure(figsize=(12, 8))

# Plot the same metric from both experiments
metric = 'train_loss/implicit_loss'
plt.plot(df1['step'][0:560], df1[metric][0:560], label='Jacobian approach')
plt.plot(df2['step'][0:560], df2[metric][0:560], label='Equilibrium propagation')

plt.xlabel('Step')
plt.ylabel('Value')
plt.title('Training loss comparison')
plt.legend()
plt.grid(True)
plt.savefig('loss_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```
